**Na√Øve Bayes Algorithm**

Naive Bayes is a probabilistic classification algorithm based on Bayes' Theorem, assuming independence among features. It is widely used for text classification (spam detection, sentiment analysis), medical diagnosis, and recommendation systems.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/Labs/ML/Social_Network_Ads.csv')
dataset.head()
```

```python
dataset.Purchased.value_counts()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
NB_classifier = GaussianNB()
NB_classifier.fit(X_train, y_train)
y_pred_NB = NB_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_NB))
```
